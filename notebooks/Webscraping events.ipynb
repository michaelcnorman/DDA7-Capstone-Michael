```python
def buildurl(item):
    try :
        return 'https://www.nashville-theatre.com/theaters' + item.find('h3').find('a')['href']
    except :
        return None

def getShows(month):
    yearlist = []
    New_URL = f'https://www.nashville-theatre.com/category_dates.php?year=2022&month={month}'
    r = requests.get(New_URL, headers=headers)
    nashville_soup = BS(r.text,'html.parser')
    shows = nashville_soup.find_all('div', {'class':'sbl2-heading'})
    for item in shows:
        show = {
        'event': item.find('h3').text,
        'venue': item.find('span',{'class':'sbl2-venue'}).text,
        'dates' : item.find('span',{'class':'sbl2-date'}).text,
        'link' : buildurl(item)
        }
        yearlist.append(show)
    return yearlist
```

```python
Shows_List = []

for monthnum in range(9,13):
    Shows_List.extend(getShows(monthnum))
```

```python
#for items, row itterows()

show_times = []
  
for index, row in Shows_df.iterrows():
    event = row['event'].replace(' ','-').lower()
    venue = row['venue']
    venue = venue_dict[venue]
    shows_url = f'https://www.nashville-theatre.com/theaters/{venue}/{event}.php'
    r = requests.get(shows_url, headers=headers)
    shows_soup = BS(r.text,'html.parser')
    shows = {
      'name' : row['event'],
      'venue': row['venue'],
      'tags' : [entry.text for entry in shows_soup.find_all('a', {'class':'tag reverse'})],
      'dates': [entry.text for entry in shows_soup.find_all('div',{'class':'col1'})],
      'time1': [entry.text for entry in shows_soup.find_all('div',{'class':'col2'})],
      'time2': [entry.text for entry in shows_soup.find_all('div',{'class':'col3'})]
            }
    show_times.append(shows)
```

```python
shows_and_dates = pd.DataFrame(show_times)
shows_and_dates
```

```python
shows_and_dates_trimmed = shows_and_dates[shows_and_dates['tags'].apply(len)>0]
shows_and_dates_trimmed
```

```python
shows_and_dates_trimmed['times'] = shows_and_dates_trimmed.time1 + shows_and_dates_trimmed.time2
Popular_Shows = shows_and_dates_trimmed.drop(columns=['time1','time2']).reset_index(drop=True)
Popular_Shows = Popular_Shows[Popular_Shows['dates'].notnull()]
Popular_Shows
```

```python
pop_dates = Popular_Shows.explode('dates', ignore_index=True)
pop_dates = pop_dates.explode('times', ignore_index=True)
pop_dates = pop_dates[~pop_dates.dates.str.contains('When', na=False)]
pop_dates = pop_dates[~pop_dates.times.str.contains('What time', na=False)]
pop_dates = pop_dates[~pop_dates.times.str.contains('-', na=False)].reset_index(drop=True)
pop_dates = pop_dates[pop_dates.times != '\xa0']
pop_dates = pop_dates.drop_duplicates(subset=['name','venue','dates','times']).reset_index(drop=True)
pop_dates
```

```python
event_categories = pop_dates[['name','tags']]
events_categories = event_categories.explode('tags', ignore_index=True)
events_categories
```
